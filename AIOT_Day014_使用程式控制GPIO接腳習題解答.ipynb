### **解答2**

本單元內教材的範例程式碼如下, 可以按照範例程式碼改成作業2要求的段落

image.png

按照上面的程式碼片段，可以很容易的加上各個ＰＷＭ參數的程式碼，執行並且觀察結果。

```
from gpiozero import PWMLED
from time import sleep

# sleepInterval 每個指令之間的等待時間
sleepInterval = 1
# 定義使用GPIO 17作為GPIO的接腳
led = PWMLED(17)
while True:
  led.value = 0.1
  sleep(sleepInterval)
  led.value = 0.3
  sleep(sleepInterval)
  led.value = 0.5
  sleep(sleepInterval)
  led.value = 0.7
  sleep(sleepInterval)
```

### 作業3

---



實際練習GPIOZero透過Button控制LED，在按鈕的過程中，觀察實際按下按鈕的次數，LED點亮的次數，是否一致。練習修改程式，讓按鈕按下是全亮，按鈕放開後是30%的亮度

```
from gpiozero import LED, Button
from singal import pause

led = LED(17)

def lightOn():
  led.value = 1
  print("Light value is 1")

def lightOff():
  led.value = 0.3
  print("Light value is 0.3")

button = Button(2)
button.when_pressed = lightOn
button.when_released = lightOff

pause()
```
