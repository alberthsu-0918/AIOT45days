- 提示: 用下列指令下載作業指定影像到colab上
`!wget -O ./test.png https://i.imgur.com/hM6uFKZ.png`


```
import imghdr
!wget -O ./test.png https://i.imgur.com/hM6uFKZ.png
img_path = './test.png'

if imghdr.what(img_path) == 'png':
  print('這是一張png格式的影像')
```
